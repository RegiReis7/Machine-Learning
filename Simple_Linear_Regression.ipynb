Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('height-weight-dataset.csv')
```

```python
df.head(
)
```

```python
df.describe()
```

```python
plt.scatter(df['Height'], df['Weight'])
plt.ylabel('Height')
plt.xlabel('Weight')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['Weight']]
Y = df['Height']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

```python
X.shape
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()

linreg.fit(X_train, Y_train)
```

```python
print("The Slope of Weight is :", linreg.coef_)
print("The Intercept of Weight is :", linreg.intercept_)
```

```python
plt.scatter(X_train, Y_train)
plt.plot(X_train, linreg.predict(X_train), color='red')
plt.ylabel('Height')
plt.xlabel('Weight')
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = linreg.predict(X_test)

mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print("Mean Squared Error :", mse)
print("R2 Score :", r2)
```

```python
plt.scatter(y_pred, Y_test)
plt.xlabel('Predicted Height')
plt.ylabel('Actual Height')
```
